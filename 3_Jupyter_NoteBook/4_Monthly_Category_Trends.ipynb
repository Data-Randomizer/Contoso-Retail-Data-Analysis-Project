## 📊 Setup: Import Libraries & Format Currency

```python
# Import tools for working with data, numbers, and charts
import pandas as pd              # For handling data tables (like Excel)
import numpy as np               # For numerical operations
import matplotlib.pyplot as plt # For creating visualizations (charts)
from matplotlib.ticker import FuncFormatter  # For customizing axis labels


# -------------------------------
# 💰 Function: Format Numbers as Money
# -------------------------------
# This function converts large numbers into a cleaner format:
# Example:
# 1,500,000 → 1.5M
# 25,000 → 25K
# 500 → 500

def money_formatter(x, _):
    # If number is 1 million or more → convert to millions (M)
    if x >= 1_000_000:
        return f'{x/1_000_000:.1f}M'
    
    # If number is 1 thousand or more → convert to thousands (K)
    elif x >= 1_000:
        return f'{x/1_000:.0f}K'
    
    # Otherwise → return the number as is
    return f'{x:.0f}'
```

## 📈 Analysis: Monthly Net Sales Trend by Category (2023)

```python
# Load dataset containing monthly net sales by category
df4 = pd.read_csv(
    "D:/Downloads/Retail_Dataset_Project/2.5_Extracted Data/4_Monthly NetSales By Category  For 2023.csv"
)


# ----------------------------------------------------------
# 🎨 Define Colors and Labels
# ----------------------------------------------------------

colors = [
    "#4C72B0",  # blue
    "#C44E52",  # red
    "#55A868",  # green
    "#8172B2",  # purple
    "#CCB974",  # gold
    "#64B5CD",  # cyan
    "#8C8C8C",  # gray
    "#E17C05"   # orange
]

# Month labels for X-axis
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Maximum value in dataset for scaling Y-axis
maxvalue = int(df4['netsalesusd'].max())

# Order categories by total net sales (largest first)
Cats = df4.groupby("categoryname")["netsalesusd"].sum().sort_values(ascending=False).index


# ----------------------------------------------------------
# 🎨 Create Line Chart
# ----------------------------------------------------------

plt.style.use("ggplot")
fig, ax = plt.subplots(figsize=(15, 8))

# Plot each category as a separate line
for Cat, color in zip(Cats, colors):
    # Filter data for the current category
    df_cat = df4[df4["categoryname"] == Cat].set_index("month")
    
    ax.plot(
        df_cat.index,           # X-axis: month numbers
        df_cat["netsalesusd"],  # Y-axis: net sales
        color=color,            # Line color
        marker="o",             # Marker for each point
        label=Cat               # Add to legend
    )


# ----------------------------------------------------------
# 🔧 Customize Axes and Labels
# ----------------------------------------------------------

# Set X-axis ticks to show month names
ax.set_xticks(df_cat.index)
ax.set_xticklabels(months, fontsize=12)

# Set Y-axis ticks at intervals of 200,000
ax.set_yticks(range(0, int(maxvalue * 1.10), 200_000))

# Format Y-axis numbers as money
ax.yaxis.set_major_formatter(FuncFormatter(money_formatter))


# ----------------------------------------------------------
# 🏷️ Add Title and Legend
# ----------------------------------------------------------

ax.set_title(
    "Net Sales For Each Month By Category",
    fontsize=17,
    fontweight="bold",
    fontfamily="serif"
)

ax.legend(title="Category", loc="upper right")

# Adjust layout to prevent overlapping
plt.tight_layout()


# ----------------------------------------------------------
# 🚀 Show Chart
# ----------------------------------------------------------

plt.show()
```

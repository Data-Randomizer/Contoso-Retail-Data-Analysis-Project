## 📊 Setup: Import Libraries & Format Currency

```python
# Import tools for working with data, numbers, and charts
import pandas as pd              # For handling data tables (like Excel)
import numpy as np               # For numerical operations
import matplotlib.pyplot as plt # For creating visualizations (charts)
from matplotlib.ticker import FuncFormatter  # For customizing axis labels


# -------------------------------
# 💰 Function: Format Numbers as Money
# -------------------------------
# This function converts large numbers into a cleaner format:
# Example:
# 1,500,000 → 1.5M
# 25,000 → 25K
# 500 → 500

def money_formatter(x, _):
    # If number is 1 million or more → convert to millions (M)
    if x >= 1_000_000:
        return f'{x/1_000_000:.1f}M'
    
    # If number is 1 thousand or more → convert to thousands (K)
    elif x >= 1_000:
        return f'{x/1_000:.0f}K'
    
    # Otherwise → return the number as is
    return f'{x:.0f}'
```

## 📈 Analysis: Net Sales for Each Year

```python
# Load the dataset containing yearly net sales
df1 = (
    pd.read_csv("D:/Downloads/Retail_Dataset_Project/2.5_Extracted Data/1_Yearly NetSales.csv")
    
    # Set the "year" column as the index (so it appears on the x-axis)
    .set_index("year")
    
    # Rename column to make it cleaner and easier to read
    .rename(columns={"netsalesusd": "Net_Sales"})
)


# --------------------------------------------
# 🎨 Create Line Chart
# --------------------------------------------

# Use a clean and visually appealing style
plt.style.use("ggplot")

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))


# Plot the data:
# - X-axis → years
# - Y-axis → net sales values
# - marker="o" → adds dots to each data point
ax.plot(df1.index, df1.values, color="blue", marker="o")


# --------------------------------------------
# 🏷️ Customize Chart Appearance
# --------------------------------------------

# Add a title to explain what the chart shows
ax.set_title(
    "What Are The Net Sales For Each Year ?",
    fontweight="bold",
    fontfamily="serif"
)

# Format the Y-axis to display values as money (K, M)
ax.yaxis.set_major_formatter(FuncFormatter(money_formatter))


# --------------------------------------------
# 🚀 Display the Chart
# --------------------------------------------

plt.show()
```

## 📊 Analysis: Top 10 Net Sales by Location (2023)

```python
# Load dataset containing net sales by location
df2 = (
    pd.read_csv("D:/Downloads/Retail_Dataset_Project/2.5_Extracted Data/2_Net Sales by Location For 2023.csv")
    
    # Set "location" as index (used for labeling bars)
    .set_index("location")
)


# ----------------------------------------------------------
# 🎨 Create Horizontal Bar Chart
# ----------------------------------------------------------

# Apply a clean visual style
plt.style.use("ggplot")

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))


# Extract the sales values (first column)
values = df2.iloc[:, 0]


# ----------------------------------------------------------
# 🎨 Apply Color Gradient (for better visual effect)
# ----------------------------------------------------------

# Normalize values (scale between min and max)
norm = plt.Normalize(values.min(), values.max())

# Create a blue color gradient based on values
colors = plt.cm.Blues((0.5 + 0.3 * norm(values)))

# Create horizontal bars
bars = ax.barh(df2.index, values, color=colors)


# ----------------------------------------------------------
# 💰 Format Axis
# ----------------------------------------------------------

# Format X-axis to display values as money (K, M)
ax.xaxis.set_major_formatter(FuncFormatter(money_formatter))


# ----------------------------------------------------------
# 🏷️ Add Labels to Each Bar
# ----------------------------------------------------------

# Loop through each bar and display its value
for bar in bars:
    width = bar.get_width()  # Get the value of the bar
    
    ax.text(
        width + values.max() * 0.01,                 # Position slightly after bar
        bar.get_y() + bar.get_height() / 2,          # Center vertically
        f"{width:,.0f}",                             # Format number with commas
        va='center',                                 # Vertical alignment
        fontsize=10
    )


# ----------------------------------------------------------
# 🔧 Final Adjustments
# ----------------------------------------------------------

# Add some space to the right so labels don’t get cut off
ax.set_xlim(0, values.max() * 1.1)

# Invert Y-axis so highest value appears at the top
ax.invert_yaxis()

# Add title
ax.set_title("Top 10 Net Sales By Location For 2023", fontsize=19)

# Adjust layout to prevent overlapping
plt.tight_layout()


# ----------------------------------------------------------
# 🚀 Show Chart
# ----------------------------------------------------------

plt.show()
```

```python
# Load dataset containing gross profit by store
df3 = (
    pd.read_csv("D:/Downloads/Retail_Dataset_Project/2.5_Extracted Data/3_Store by GrossProfit For 2023.csv")
    
    # Set "store" as the index
    .set_index("store")
    
    # Sort by gross profit descending to get top performers
    .sort_values("grossprofit", ascending=False)
    
    # Select the top 10 stores
    .head(10)
    
    # Sort again ascending for horizontal bar chart display
    .sort_values("grossprofit")
)

# Clean up store names by removing repetitive text
df3.index = df3.index.str.replace("Contoso Store ", "", regex=False)


# ----------------------------------------------------------
# 🎨 Create Horizontal Bar Chart
# ----------------------------------------------------------

plt.style.use("ggplot")
fig, ax = plt.subplots(figsize=(10, 6))

# Extract gross profit values
values = df3.iloc[:, 0]

# Normalize values for gradient coloring
norm = plt.Normalize(values.min(), values.max())
colors = plt.cm.Blues(0.3 + 0.4 * norm(values))

# Draw horizontal bars
bars = ax.barh(df3.index, values, color=colors)


# ----------------------------------------------------------
# 💰 Format Axis
# ----------------------------------------------------------

# Format X-axis numbers as money (K, M)
ax.xaxis.set_major_formatter(FuncFormatter(money_formatter))


# ----------------------------------------------------------
# 🏷️ Add Value Labels on Bars
# ----------------------------------------------------------

for bar in bars:
    width = bar.get_width()
    ax.text(
        width + values.max() * 0.01,       # Slightly after the bar
        bar.get_y() + bar.get_height() / 2, # Center vertically
        f"{width:,.0f}",                   # Format with commas
        va='center',
        fontsize=10
    )


# ----------------------------------------------------------
# 🔧 Final Adjustments
# ----------------------------------------------------------

# Add extra space on the X-axis so labels fit
ax.set_xlim(0, values.max() * 1.1)

# Add title
ax.set_title("Top 10 Net Sales By Store For 2023", fontsize=19)

# Adjust layout to prevent overlaps
plt.tight_layout()


# ----------------------------------------------------------
# 🚀 Show Chart
# ----------------------------------------------------------

plt.show()
```

## 📈 Analysis: Monthly Net Sales Trend by Category (2023)

```python
# Load dataset containing monthly net sales by category
df4 = pd.read_csv(
    "D:/Downloads/Retail_Dataset_Project/2.5_Extracted Data/4_Monthly NetSales By Category  For 2023.csv"
)


# ----------------------------------------------------------
# 🎨 Define Colors and Labels
# ----------------------------------------------------------

colors = [
    "#4C72B0",  # blue
    "#C44E52",  # red
    "#55A868",  # green
    "#8172B2",  # purple
    "#CCB974",  # gold
    "#64B5CD",  # cyan
    "#8C8C8C",  # gray
    "#E17C05"   # orange
]

# Month labels for X-axis
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Maximum value in dataset for scaling Y-axis
maxvalue = int(df4['netsalesusd'].max())

# Order categories by total net sales (largest first)
Cats = df4.groupby("categoryname")["netsalesusd"].sum().sort_values(ascending=False).index


# ----------------------------------------------------------
# 🎨 Create Line Chart
# ----------------------------------------------------------

plt.style.use("ggplot")
fig, ax = plt.subplots(figsize=(15, 8))

# Plot each category as a separate line
for Cat, color in zip(Cats, colors):
    # Filter data for the current category
    df_cat = df4[df4["categoryname"] == Cat].set_index("month")
    
    ax.plot(
        df_cat.index,           # X-axis: month numbers
        df_cat["netsalesusd"],  # Y-axis: net sales
        color=color,            # Line color
        marker="o",             # Marker for each point
        label=Cat               # Add to legend
    )


# ----------------------------------------------------------
# 🔧 Customize Axes and Labels
# ----------------------------------------------------------

# Set X-axis ticks to show month names
ax.set_xticks(df_cat.index)
ax.set_xticklabels(months, fontsize=12)

# Set Y-axis ticks at intervals of 200,000
ax.set_yticks(range(0, int(maxvalue * 1.10), 200_000))

# Format Y-axis numbers as money
ax.yaxis.set_major_formatter(FuncFormatter(money_formatter))


# ----------------------------------------------------------
# 🏷️ Add Title and Legend
# ----------------------------------------------------------

ax.set_title(
    "Net Sales For Each Month By Category",
    fontsize=17,
    fontweight="bold",
    fontfamily="serif"
)

ax.legend(title="Category", loc="upper right")

# Adjust layout to prevent overlapping
plt.tight_layout()


# ----------------------------------------------------------
# 🚀 Show Chart
# ----------------------------------------------------------

plt.show()
```

## 📈 Interactive: Monthly Net Sales by Selected Categories (2023)

```python
# Load dataset containing monthly net sales by category
df4 = pd.read_csv(
    "D:/Downloads/Retail_Dataset_Project/2.5_Extracted Data/4_Monthly NetSales By Category  For 2023.csv"
)


# ----------------------------------------------------------
# 🎨 Define Colors and Month Labels
# ----------------------------------------------------------

colors = [
    "#4C72B0",  # blue
    "#C44E52",  # red
    "#55A868",  # green
    "#8172B2",  # purple
    "#CCB974",  # gold
    "#64B5CD",  # cyan
    "#8C8C8C",  # gray
    "#E17C05"   # orange
]

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Maximum value in dataset for scaling Y-axis
maxvalue = int(df4['netsalesusd'].max())


# ----------------------------------------------------------
# 📋 Show Available Categories
# ----------------------------------------------------------

category_unique = df4.groupby("categoryname")["netsalesusd"].sum().sort_values(ascending=False).index

print("Available categories to compare:")
for z in category_unique:
    print("-", z)


# ----------------------------------------------------------
# 🖊️ User Input: Select Categories to Plot
# ----------------------------------------------------------

Cats = []

# Ask user how many categories they want to compare
while True:
    try:
        plots = int(input("How many categories do you want to compare? ").strip())
        if plots > 0:
            break
        else:
            print("Enter a number greater than 0!")
    except ValueError:
        print("Please enter a valid number!")

# Ask user to type each category
for i in range(plots):
    cat_input = input(f"Category {i+1}: ").strip()
    
    while cat_input not in category_unique:
        print("Category not found, try again!")
        cat_input = input(f"Category {i+1}: ").strip()
    
    Cats.append(cat_input)


# ----------------------------------------------------------
# 🎨 Create Line Chart for Selected Categories
# ----------------------------------------------------------

plt.style.use("ggplot")
fig, ax = plt.subplots(figsize=(15, 8))

for Cat, color in zip(Cats, colors):
    df_cat = df4[df4["categoryname"] == Cat].set_index("month")
    
    ax.plot(
        df_cat.index,
        df_cat["netsalesusd"],
        color=color,
        marker="o",
        label=Cat
    )


# ----------------------------------------------------------
# 🔧 Customize Axes, Title, and Legend
# ----------------------------------------------------------

ax.set_xticks(df_cat.index)
ax.set_xticklabels(months, fontsize=12)

ax.set_yticks(range(0, int(maxvalue * 1.10), 200_000))
ax.yaxis.set_major_formatter(FuncFormatter(money_formatter))

ax.set_title(
    "Net Sales For Each Month By Selected Categories",
    fontsize=17,
    fontweight="bold",
    fontfamily="serif"
)

ax.legend(title="Category", loc="upper right")

plt.tight_layout()


# ----------------------------------------------------------
# 🚀 Show Chart
# ----------------------------------------------------------

plt.show()
```

## 📈 Analysis: Cumulative Gross Profit for Each Store (2024)

```python
# Load dataset containing gross profit by store over time
df5 = pd.read_csv(
    "D:/Downloads/Retail_Dataset_Project/2.5_Extracted Data/5_Top 5 Stores GrossProfit By Date For 2024.csv"
)


# ----------------------------------------------------------
# 🔧 Convert Dates and Prepare Subplots
# ----------------------------------------------------------

# Convert "orderdate" column to datetime format
df5["orderdate"] = pd.to_datetime(df5["orderdate"])

# Count number of unique stores to create one subplot per store
count = len(df5["storename"].unique())
fig, ax = plt.subplots(count, 1, figsize=(15, 20))


# ----------------------------------------------------------
# 🎨 Define Colors for Lines
# ----------------------------------------------------------

colors = [
    "#4C72B0",
    "#C44E52",
    "#55A868",
    "#8172B2",
    "#CCB974",
    "#64B5CD",
    "#8C8C8C",
    "#E17C05"
]


# ----------------------------------------------------------
# 🖊️ Plot Cumulative Gross Profit for Each Store
# ----------------------------------------------------------

# Group data by store
for x, (store, df_store) in enumerate(reversed(list(df5.groupby("storename")))):
    
    color = colors[x % len(colors)]  # pick color for this store
    
    # Sort data by date
    df_store = df_store.sort_values("orderdate").copy()
    df_store.set_index("orderdate", inplace=True)
    
    # Calculate cumulative gross profit
    y = df_store["grossprofit"].cumsum()
    
    # Plot cumulative line on subplot
    ax[x].plot(
        df_store.index,
        y,
        color=color,
        label=store
    )
    
    # Add title for each subplot
    ax[x].set_title(
        f"Cumulative Gross Profit for {store}",
        fontsize=14,
        fontweight="bold",
        fontfamily="serif"
    )
    
    # Add legend
    ax[x].legend(title="Store", loc="upper left")
    
    # Format Y-axis as money and add grid
    ax[x].yaxis.set_major_formatter(FuncFormatter(money_formatter))
    ax[x].grid(True)


# Adjust layout to prevent overlap
plt.tight_layout()

# Show all charts
plt.show()
```

## 📊 Setup: Import Libraries & Format Currency

```python
# Import tools for working with data, numbers, and charts
import pandas as pd              # For handling data tables (like Excel)
import numpy as np               # For numerical operations
import matplotlib.pyplot as plt # For creating visualizations (charts)
from matplotlib.ticker import FuncFormatter  # For customizing axis labels


# -------------------------------
# 💰 Function: Format Numbers as Money
# -------------------------------
# This function converts large numbers into a cleaner format:
# Example:
# 1,500,000 → 1.5M
# 25,000 → 25K
# 500 → 500

def money_formatter(x, _):
    # If number is 1 million or more → convert to millions (M)
    if x >= 1_000_000:
        return f'{x/1_000_000:.1f}M'
    
    # If number is 1 thousand or more → convert to thousands (K)
    elif x >= 1_000:
        return f'{x/1_000:.0f}K'
    
    # Otherwise → return the number as is
    return f'{x:.0f}'
```

## 📈 Analysis: Net Sales for Each Year

```python
# Load the dataset containing yearly net sales
df1 = (
    pd.read_csv("D:/Downloads/Retail_Dataset_Project/2.5_Extracted Data/1_Yearly NetSales.csv")
    
    # Set the "year" column as the index (so it appears on the x-axis)
    .set_index("year")
    
    # Rename column to make it cleaner and easier to read
    .rename(columns={"netsalesusd": "Net_Sales"})
)


# --------------------------------------------
# 🎨 Create Line Chart
# --------------------------------------------

# Use a clean and visually appealing style
plt.style.use("ggplot")

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))


# Plot the data:
# - X-axis → years
# - Y-axis → net sales values
# - marker="o" → adds dots to each data point
ax.plot(df1.index, df1.values, color="blue", marker="o")


# --------------------------------------------
# 🏷️ Customize Chart Appearance
# --------------------------------------------

# Add a title to explain what the chart shows
ax.set_title(
    "What Are The Net Sales For Each Year ?",
    fontweight="bold",
    fontfamily="serif"
)

# Format the Y-axis to display values as money (K, M)
ax.yaxis.set_major_formatter(FuncFormatter(money_formatter))


# --------------------------------------------
# 🚀 Display the Chart
# --------------------------------------------

plt.show()
```

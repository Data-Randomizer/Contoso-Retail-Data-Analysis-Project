## 📊 Setup: Import Libraries & Format Currency

```python
# Import tools for working with data, numbers, and charts
import pandas as pd              # For handling data tables (like Excel)
import numpy as np               # For numerical operations
import matplotlib.pyplot as plt # For creating visualizations (charts)
from matplotlib.ticker import FuncFormatter  # For customizing axis labels


# -------------------------------
# 💰 Function: Format Numbers as Money
# -------------------------------
# This function converts large numbers into a cleaner format:
# Example:
# 1,500,000 → 1.5M
# 25,000 → 25K
# 500 → 500

def money_formatter(x, _):
    # If number is 1 million or more → convert to millions (M)
    if x >= 1_000_000:
        return f'{x/1_000_000:.1f}M'
    
    # If number is 1 thousand or more → convert to thousands (K)
    elif x >= 1_000:
        return f'{x/1_000:.0f}K'
    
    # Otherwise → return the number as is
    return f'{x:.0f}'
```

## 📊 Analysis: Top 10 Net Sales by Location (2023)

```python
# Load dataset containing net sales by location
df2 = (
    pd.read_csv("D:/Downloads/Contoso Retail Data Analysis Project/2.5_Extracted Data/2_Net Sales by Location For 2023.csv")
    
    # Set "location" as index (used for labeling bars)
    .set_index("location")
)


# ----------------------------------------------------------
# 🎨 Create Horizontal Bar Chart
# ----------------------------------------------------------

# Apply a clean visual style
plt.style.use("ggplot")

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))


# Extract the sales values (first column)
values = df2.iloc[:, 0]


# ----------------------------------------------------------
# 🎨 Apply Color Gradient (for better visual effect)
# ----------------------------------------------------------

# Normalize values (scale between min and max)
norm = plt.Normalize(values.min(), values.max())

# Create a blue color gradient based on values
colors = plt.cm.Blues((0.5 + 0.3 * norm(values)))

# Create horizontal bars
bars = ax.barh(df2.index, values, color=colors)


# ----------------------------------------------------------
# 💰 Format Axis
# ----------------------------------------------------------

# Format X-axis to display values as money (K, M)
ax.xaxis.set_major_formatter(FuncFormatter(money_formatter))


# ----------------------------------------------------------
# 🏷️ Add Labels to Each Bar
# ----------------------------------------------------------

# Loop through each bar and display its value
for bar in bars:
    width = bar.get_width()  # Get the value of the bar
    
    ax.text(
        width + values.max() * 0.01,                 # Position slightly after bar
        bar.get_y() + bar.get_height() / 2,          # Center vertically
        f"{width:,.0f}",                             # Format number with commas
        va='center',                                 # Vertical alignment
        fontsize=10
    )


# ----------------------------------------------------------
# 🔧 Final Adjustments
# ----------------------------------------------------------

# Add some space to the right so labels don’t get cut off
ax.set_xlim(0, values.max() * 1.1)

# Invert Y-axis so highest value appears at the top
ax.invert_yaxis()

# Add title
ax.set_title("Top 10 Net Sales By Location For 2023", fontsize=19)

# Adjust layout to prevent overlapping
plt.tight_layout()


# ----------------------------------------------------------
# 🚀 Show Chart
# ----------------------------------------------------------

plt.show()
```
